```python
from PIL import Image

def load_image(filename):
    # Open image file
    img = Image.open(filename)
    
    # Convert image to RGB mode (if not already in RGB mode)
    img = img.convert("RGB")
    
    # Get image dimensions
    width, height = img.size
    
    # Initialize 2D array to store pixel data
    pixel_data = [[(0, 0, 0) for _ in range(width)] for _ in range(height)]
    
    # Iterate through each pixel of the image and store pixel data in the array
    for y in range(height):
        for x in range(width):
            pixel_data[y][x] = img.getpixel((x, y))
    
    return pixel_data

# Example usage
filename = "maxresdefault.jpg"
image_data = load_image(filename)
print("Image loaded successfully.")
print("Image dimensions:", len(image_data), "x", len(image_data[0]))
```
